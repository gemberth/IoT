# 3ro.  Graficar la regresión Lineal

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from PIL import Image


# Leer el archivo Excel
datos = pd.read_excel('EMPRESAS.xls')

# Obtener las variables independientes (X) y dependiente (y)
X = datos['NUMEMPLEA'].values.reshape(-1, 1)
y = datos['NUMCLIENTES'].values
datos.head()

# Crear un objeto de regresión lineal
regresion_lineal = LinearRegression()

# Entrenar el modelo de regresión lineal
regresion_lineal.fit(X, y)

# Realizar una predicción
x_prediccion = X
y_prediccion = regresion_lineal.predict(X)

# Crear la gráfica
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, y_prediccion, color='red', label='Regresión lineal')
plt.title('Regresión lineal')
plt.xlabel('Número de Clientes')
plt.ylabel('Facturación')
plt.legend()

# Agregar la marca de agua
#marca_agua = Image.open('gemberth.png')
#plt.imshow(marca_agua, extent=(0, 50, 0, 30), alpha=0.3, zorder=0)

# Mostrar la gráfica
plt.show()
```

# 5to Generar La recta de los valores de aprendizaje (20 números aleatorios )

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generar 20 números aleatorios en el rango de 0 a 10
X_aleatorio = np.random.uniform(0, 10, size=(20, 1))

# Crear los valores correspondientes de la recta de aprendizaje
y_aprendizaje = 2 * X_aleatorio + 3

# Crear el objeto de regresión lineal
regresion_lineal = LinearRegression()

# Entrenar el modelo de regresión lineal con los datos de aprendizaje
regresion_lineal.fit(X_aleatorio, y_aprendizaje)

# Obtener los coeficientes de la recta de regresión lineal
pendiente = regresion_lineal.coef_[0]
intercepto = regresion_lineal.intercept_

# Crear los valores correspondientes de la recta de regresión lineal
y_prediccion = pendiente * X_aleatorio + intercepto

# Graficar los puntos de aprendizaje y la recta de regresión lineal
plt.scatter(X_aleatorio, y_aprendizaje, color='blue', label='Valores de aprendizaje')
plt.plot(X_aleatorio, y_prediccion, color='red', label='Recta de regresión lineal')
plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')
plt.title('Recta de los valores de aprendizaje')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from PIL import Image

# Generar 20 números aleatorios en el rango de 0 a 10
X_aleatorio = np.random.uniform(0, 10, size=(20, 1))

# Crear los valores correspondientes de la recta de aprendizaje
y_aprendizaje = 2 * X_aleatorio + 3

# Crear el objeto de regresión lineal
regresion_lineal = LinearRegression()

# Entrenar el modelo de regresión lineal con los datos de aprendizaje
regresion_lineal.fit(X_aleatorio, y_aprendizaje)

# Obtener los coeficientes de la recta de regresión lineal
pendiente = regresion_lineal.coef_[0]
intercepto = regresion_lineal.intercept_

# Crear los valores correspondientes de la recta de regresión lineal
y_prediccion = pendiente * X_aleatorio + intercepto

# Graficar los puntos de aprendizaje y la recta de regresión lineal
plt.scatter(X_aleatorio, y_aprendizaje, color='blue', label='Valores de aprendizaje')
plt.plot(X_aleatorio, y_prediccion, color='red', label='Recta de regresión lineal')
plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')
plt.title('Recta de los valores de aprendizaje')

# Agregar la marca de agua
marca_agua = Image.open('gemberth.png')
plt.imshow(marca_agua, extent=(0, 30, 0, 20), alpha=0.3, zorder=0)

# Mostrar la gráfica
plt.legend()
plt.show()
```
